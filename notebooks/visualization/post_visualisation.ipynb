```python
import pandas as pd
df = pd.read_csv("/Users/macbook/house-price-prediction/notebooks/dimensionality reduction/feature_selection_train.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 8))
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, square=True)
plt.title("Matrice de Corrlation (features + SalePrice)")
plt.show()
```

```python
k = 10
cols = corr_matrix.nlargest(k, "SalePrice")["SalePrice"].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(
    cm,
    cbar=True,
    annot=True,
    square=True,
    fmt=".2f",
    annot_kws={"size": 10},
    yticklabels=cols.values,
    xticklabels=cols.values,
)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['SalePrice'], kde=True, bins=30, color='skyblue')
plt.title("Distribution de SalePrice")
plt.xlabel("SalePrice")
plt.ylabel("Frquence")
plt.show()
```

```python
# Slection des variables les plus corrles avec SalePrice
top_corr_features = corr_matrix['SalePrice'].abs().sort_values(ascending=False)[1:6].index

sns.pairplot(df[top_corr_features.tolist() + ['SalePrice']], diag_kind='kde')
plt.suptitle("Paires de variables corrles avec SalePrice", y=1.02)
plt.figure(figsize=(5, 3))
plt.show()

```
