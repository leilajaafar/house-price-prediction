```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
selected_columns = [
    "OverallQual",
    "GarageCars",
    "GarageArea",
    "ExterQual",
    "GarageFinish",
    "KitchenQual",
    "BsmtQual",
    "YearBuilt",
    "TotalBsmtSF",
    "GrLivArea",
    "1stFlrSF",
    "FullBath",
    "YearRemodAdd",
    "Fireplaces",
    "HeatingQC",
    "TotRmsAbvGrd",
    "MasVnrArea",
    "BsmtFinSF1",
    "BsmtExposure",
    "LotFrontage",
    "WoodDeckSF",
    "OpenPorchSF",
    "Neighborhood"
]
```

```python
df_train = pd.read_csv(
    r"C:\Users\leila\Downloads\pfaproject\house-price-prediction\data\raw data\train.csv",
    usecols=selected_columns + ['SalePrice']
)
```

```python
df_train.head()
```

```python
df_train.columns
```

```python
df_train.info()
```

```python
### 1. Traitement des valeurs manquantes selon la description ###

# Variables catégorielles avec "NA" dans la description
na_categorical = {
    "BsmtQual": "NA",  # No Basement
    "BsmtExposure": "NA",  # No Basement
    "GarageFinish": "NA",  # No Garage
}

for col, na_value in na_categorical.items():
    df_train[col] = df_train[col].fillna(na_value)


# Variables numériques
df_train["LotFrontage"] = df_train.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
df_train["MasVnrArea"] = df_train["MasVnrArea"].fillna(0)  # Si pas de maçonnerie

### 3. Vérification finale ###
print("Valeurs manquantes restantes:")
print(df_train.isnull().sum().sum())  
```

```python
df_train.info()
```

```python
# ---------------------------------------------------------------
# Ordinal Encoding (manual) 
# ---------------------------------------------------------------

ordinal_mapping = {
    'ExterQual': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1},
    'BsmtQual': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0},
    'BsmtExposure': {'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1, 'NA': 0},
    'HeatingQC': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1},
    'KitchenQual': {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1},
    'GarageFinish': {'Fin': 3, 'RFn': 2, 'Unf': 1, 'NA': 0}
}

for col, mapping in ordinal_mapping.items():
    if col in df_train.columns:
        df_train[col] = df_train[col].map(mapping)

# ---------------------------------------------------------------
# 2. One-Hot Encoding (variables nominales)
# ---------------------------------------------------------------
onehot_cols = ['Neighborhood']
df_train = pd.get_dummies(df_train, columns=onehot_cols, drop_first=True)
```

```python
# Sélection juste des colonnes numériques
#numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
numeric_cols = [
    col for col in df_train.select_dtypes(include=['int64', 'float64']).columns 
    if col not in ['Id', 'SalePrice']
]
scaler = StandardScaler()
# Appliquer la standardisation
df_standardized = df_train.copy()
df_standardized[numeric_cols] = scaler.fit_transform(df_train[numeric_cols])
```

```python
#Sauvegarder le résultat
df_standardized.to_csv("feature_selection_train.csv", index=False)
```

```python
# Columns to keep:
cols_to_keep = [
    'LotFrontage', 'OverallQual', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea',
    'ExterQual', 'BsmtQual', 'BsmtExposure', 'BsmtFinSF1', 'TotalBsmtSF',
    'HeatingQC', '1stFlrSF', 'GrLivArea', 'FullBath', 'KitchenQual',
    'TotRmsAbvGrd', 'Fireplaces', 'GarageFinish', 'GarageCars',
    'GarageArea', 'WoodDeckSF', 'OpenPorchSF'
]

# ---------------------------------------------------------------
# 1. Fill NA in categorical columns from your selection if needed
# From your cols, categorical with possible NA: BsmtQual, BsmtExposure, KitchenQual, GarageFinish, ExterQual, HeatingQC
# Use 'NA' as string for missing values in these ordinal-like categories
cat_na_cols = ['BsmtQual', 'BsmtExposure', 'KitchenQual', 'GarageFinish', 'ExterQual', 'HeatingQC']
df_test[cat_na_cols] = df_test[cat_na_cols].fillna('NA')

# ---------------------------------------------------------------
# 2. Impute LotFrontage using neighborhood medians (if Neighborhood is present)
# Utiliser les médianes calculées sur df_train pour LotFrontage
neighborhood_medians = df_test.groupby('Neighborhood')['LotFrontage'].median()
df_test['LotFrontage'] = df_test.apply(
    lambda row: neighborhood_medians[row['Neighborhood']] if pd.isna(row['LotFrontage']) else row['LotFrontage'], 
    axis=1
)

# ---------------------------------------------------------------
# 3. Fill numeric columns NAs with 0 (from your list)
numeric_zero_cols = ['MasVnrArea', 'BsmtFinSF1', 'TotalBsmtSF', 'GarageCars', 'GarageArea']
df_test[numeric_zero_cols] = df_test[numeric_zero_cols].fillna(0)

# ---------------------------------------------------------------
# 4. If no basement (BsmtQual == 'NA'), basement-related numeric columns = 0
bsmt_related = ['BsmtFinSF1', 'TotalBsmtSF']
df_test.loc[df_test['BsmtQual'] == 'NA', bsmt_related] = 0

# ---------------------------------------------------------------
# 5. If no garage (GarageFinish == 'NA'), garage-related numeric columns = 0
garage_related = ['GarageCars', 'GarageArea']
df_test.loc[df_test['GarageFinish'] == 'NA', garage_related] = 0

# ---------------------------------------------------------------
# 6. Final check of missing values on these columns
missing_summary = df_test[cols_to_keep].isnull().sum()
print("Missing values in df_test (selected columns):")
print(missing_summary)
```
